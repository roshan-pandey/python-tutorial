## What are Variables?
- They hold some value, they are containers. Like we have jars at out home and we store things in them.
- In python we don't have to specifically mention the data type of the variable. Based on the value assigned to the variable data type is determined.

```python
x = 3 # integer variable (int)
y = "How are you?" # string variable (str)
print("This is the value of x:",x)
print("This is the value of y:",y)
```

```python
# to check the type of a variable we use type(variable_name)
print("This is the type of x:", type(x))
print("This is the type of y:", type(y))
```

## Type Casting
- When we change data type of a variable, it is know as casting or type casting.


```python
x = 3
print("Value of x is:", x, "and it's type is:", type(x)) 
y = str(x)
print("Value of y is:", y, "and it's type is:", type(y))
z = float(x)
print("Value of y is:", z, "and it's type is:", type(z))
```

- Variables are case sensitive

```python
r = "Hello"
R = "Programming Cradle"
print(r)
print(R)
```

## Rules to follow while naming a variable.
- Variable name should start with a letter or underscore
- Variable name cannot start with a number
- Variable name can contain only alpha-numerics and underscores (A-z, 0-9, & _ )

## Variable Naming Convention in Python
- When there are more than one words in a variable they should be separated by underscore eg. first_name, second_name. 
- We should use lower case characters.


```python
# some examples for variable name

first_name = "John"
last_name = "Doe"
age = 30
home_permanent_address = "Test address"
```

## Assigning values to the variable

```python
x = y = z = 50 # this way you can assign same value to multiple variables
print(x)
print(y)
print(z)
```

```python
x, y, z = 30, 60, 90 # assign different values to different variables
print(x)
print(y)
print(z)
```

```python
# if you have values in a list and you want to assign each element/value to different variables it is called as unpacking
num_list = [10, 100, 1000]
x, y, z = num_list
print(x)
print(y)
print(z)
```
