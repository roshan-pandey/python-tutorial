## Conditional Statements in Python

#### if statement, if-else statement, if-elif-else.

### Comparision operatores used in conditional statements
- Equals-> a == b <br>
- Not Equals-> a != b <br>
- Less than-> a < b <br>
- Greater than-> a > b <br>
- Less than or equal to-> a <= b <br>
- Greater than or equal to-> a >= b <br>

```python
# if statement

a = 5
b = 10
if b/a == 2:
  print("b is double of a")
```

```python
# if-else statement

if a > b:
  print("a wins")
else:
  print("b wins")
```

```python
# if-elif-else statement

if a == b:
  print("Match tie")
elif a < b:
  print("b wins")
else:
  print("a wins")
```

```python
# nested if
if a < b:
  if b/a:
    print("Divisible")
  else:
    print("Non divisible")
else:
  print("Never Mind")
```

```python
# pass keyword

if a <= b:
  pass # Do nothing...
```
