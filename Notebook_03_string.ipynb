### Strings in Python

- String in python is a sequence of characters. Each alphabet, number, or symbol is a character, eg. "Hello" --> here h, e,  l, l, o are characters and "Hello" is a string.
- String is always enclosed in double quotes(" ") or single quotes(' ').
- Multiline string is enclosed in 3 double or single quotes (""" """ or ''' ''')

```python
str1 = "Hello"
str2 = ' Programming Cardle'
print(str1+str2)

print(str1+","+str2+'!')
```

```python
multiline_str = '''Hello! How are you?
Where are you from?
Do we know each other?'''

print(multiline_str)
```

### Indexing Strings

- As mentioned before, Strings are sequence of chars. So, we can access each chars separately using index. eg. str1[1]

```python
print("First Char in str1 is:", str1[0])
```

### Slicing of Strings

- We can extract some part of the string using slicing. eg str2[3:6] --> this will extract char from index 3 to 5 (includes starting index, excludes ending index)

```python
str3 = "Learning Python"
print(str3[1:5]) # syntax: variable_name[starting_index, ending_index + 1]
```

```python
# Extracting sub string from start till some specific index.
print(str3[:8])

# Extracting sub string from some specific index till last. 
print(str3[9:])
```

### Negative indexing in python

- When you want to slice a string or extract a specific char from end you can use negative indexing. eg. str3[-6:]

```python
print(str3[-6:])
```

### Important String Functions

- There are many pre-defined string manipulation functions. We will cover important ones. If you want to check out all the functions you can vist [here](https://docs.python.org/3/library/stdtypes.html#string-methods)

```python
print(str3.upper()) # all chars to upper case
print("-"*30)

print(str3.lower()) # all chars to lower case
print("-"*30)

str4 = " String with Space in front and back "
print(str4)
print(str4.strip()) # removes white space from start and end
print("-"*30)

print(str3.replace("Python", "Programming")) # replaces certain char or string with another char or string
print("-"*30)

str5 = "What, That, When, Then"
print(str5.split(",")) # splits the given string at specified char and returns a list
print("-"*30)

list1 = ['What', ' That', ' When', ' Then']
print(",".join(list1)) # joins each element of the list and returns a string.
```

### String Concatenation in Python

- We can use + operator to concate 2 or more strings. eg. str1+str2

```python
str6 = "Programming"
str7 = "Cradle"
result1 = str6+str7

print(result1)
print("-"*30)

result2 = str6+" "+str7
print(result2)
```

### Escape charachers

- To ignore the special meaning of the symbol or char we use escape character i.e., \ 

There are several other excape chars in python check [here](https://docs.python.org/2.0/ref/strings.html)

```python
str8 = "There is a \"dog\" in the backyard." # here we want to use double quotes as simple string rather than special char. so, we will use \
print(str8)
```
