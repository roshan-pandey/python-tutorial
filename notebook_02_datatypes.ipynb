# Data Type

- Data type of the variable is the type of the data it stores, i.e., str, int, float, etc.

### Python has following data types

**Text Type**:	str </br>
**Numeric Types**:	int, float, complex </br>
**Sequence Types**:	list, tuple, range </br>
**Mapping Type**:	dict </br>
**Set Types**:	set </br>
**Boolean Type**:	bool </br>


```python
a = "Hello, Programming Cradle"
b = 3
c = 3.3
d = 3j
e = ["hello", "hi", "hola"]
f = ("hello", "hi", "hola")
j = range(0,6) #syntax: range(start, end, step_size) by default step_size = 1
k = {"key1": "Value1", "key2": "Value2"} #syntax: {key:value} 
l = {1, 2, 3, 4}
m = True
n = False
```

### Type Casting in Python

Changing the type of the variable is known as casting
- int(): to change variable type to integer.
- float(): to change variable type to float.
- str(): to change variable type to string.

```python
x = 3
print("Value of x is:", x, "and type is:",type(x))

y = float(x)
print("Value of y is:", y, "and type is:", type(y))

z = str(x)
print("Value of z is:", z, "and type is:", type(z))

w = int(z)
print("Value of w is:", w, "and type is:", type(w))
```

```python
p = 1.2
q = int(p)
print(q)

print("-"*10)
r = str(p)
print(r)
```
