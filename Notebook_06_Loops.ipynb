## Looping statements in Python
There are 2 types of looping statements in python:
- while loop
- for loop

### While Loop

```python
a = 0
while a < 5:
  a = a + 1
  print(a)
```

```python
# if we want to come out of the loop before it completes all the iterations we can use "break" command

b = 0
while b < 10:
  b += 1
  print(b)
  if b == 5:
    break
```

```python
# if we want to skip some condition... we can use "continue" command

c = 0
while c < 6:
  c += 1
  if c == 2 or c == 4:
    continue
  else:
    print(c)
```

```python
# we can also use "else" with while statement

d = 0
while d < 4:
  d += 1
  print(d)
else:
  print("We have reached end of the queue")
```

### For Loop

```python
# For loop and while loop can be used interchangeably which ever you are comfortable with.

for i in range(0, 5, 1): 
  print(i)

# NOTE: As you can see last value is 1 less than the upper limit... So, Keep this in mind... 
# You can use break, continue, and else with for loop as well...
```

```python
# Looping over list...
l = ["Blue", 2, 4, "ten", 9.5]
for i in l:
  print(i)

print("-"*30)
# we can also loop over string in the same way...
s = "Something"
for i in s:
  print(i)
```

```python
# Nested Loop...
# Both while and for loop can be nested...
l1 = [1, 2, 3]
l2 = ["a", "b", "c"]
for i in l1:
  for j in l2:
    print(j + " in iteration " + str(i))
```
