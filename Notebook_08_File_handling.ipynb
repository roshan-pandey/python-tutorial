## File Handling in Python

For Data Science we mostly uses packages like Pandas and Dask for file reading or writing. But in several instances you might want to use python file handling constructs. 

open(): To open a file, we use this function. It takes 2 parameters file name and mode in which the file has to be opened. <br>

Modes:
- "r" - Read - By Default... Opens a file for reading and throws error if the file does not exist. <br>
- "w" - Write - Opens a file for writing and creates the file if it does not exist. <br>
- "a" - Append - Opens a file for appending and creates the file if it does not exist. <br> 
- "x" - Create - Creates the specified file and returns an error if the file exists. <br>


NOTE: You can also specipy whether the file has to be handled as text or binary(for image data).

```python
# Creating a file... 

f = open("sample_file.txt", "x") # create a file with name sample_file
```

```python
# Writing to a file...

fw = open("sample_file.txt", "w")
fw.write("I am writing this bcoz I have to write something... :p \n")
fw.close()
```

```python
# Reading a file...

fr = open("sample_file.txt", "rt")
txt = fr.read()
print(txt)
```

```python
txt.replace("bcoz", "because")
```

NOTE: read() takes out all the data at once but if you want to read line by line you can use readline()

```python
fr.close() # always close the file handler... Just like you close the door... :D
```

```python
# Appending to a file...

fa = open("sample_file.txt", "a")
fa.write("Let's add another boring line to the file...")
fa.close()
```

```python
fr = open("sample_file.txt", "r")
print(fr.readline())
print(fr.readline())
fr.close()
```

```python
# You can also use for loop to go through the file content line by line

f = open("sample_file.txt", "r")
for i in f:
  print("-"*10+ i)
f.close()
```

```python
# Deleting a file...
import os
os.remove("sample_file.txt")
```
