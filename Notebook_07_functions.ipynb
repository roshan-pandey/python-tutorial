## Functions in Python

Functions are snippet of code wrapped in a block which can only be executed on demand(function call).

```python
# function to add 2 numbers
def add_num():
  a = 7
  b = 17
  res = a + b
  return res
```

```python
# Function Call
result = add_num()
print(result)
```

### Parameters in funtion
- You can pass values in a function to make it work for different values.

```python
# Function with parameters
def add_val(a, b):
  return a+b
```

```python
result = add_val(39, -15)
print(result)
```

You can pass as many number of parameters as you like and if you don't know how many parameters you want to pass, you can add * before the parameter name.

```python
def random_func(*some_param):
  print("I saw " + some_param[0] +" sitting in a " +some_param[1] +" shop")
```

```python
random_func("Susie", "shoeshine")
```

### Keyword parameters
If there are many parameters to pass and we don't want to mess up the order we use keyword parameters

```python
def many_param(p2, p4, p1, p3):
  print("Driver Standings are:")
  print("Position 1: "+p1)
  print("Position 2: "+p2)
  print("Position 3: "+p3)
  print("Position 4: "+p4)
```

```python
many_param(p1="Ham", p2="Ver", p3="Bot", p4="Nor")
```

### Default Parameters
When some field is optional it can take default value. 

```python
def default_func(fname, lname, mname=""): #default parameter is always last parameter
  print(fname+" "+mname+" "+lname)
```

```python
default_func("John", "Doe", "Smith")
default_func("John", "Doe")
```

## Recursive Functions
Funtion that calls itself is known as recursive function

```python
def rec_fun(a): # recursive function to add number 1 to 10
  if a > 9:
    return a
  else:
    return a + rec_fun(a+1)
```

```python
res = rec_fun(1)
print(res)
```
